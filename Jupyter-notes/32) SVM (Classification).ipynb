```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
data = pd.read_csv(r"D:\Data Science\Datasets\placement-dataset.csv")
data.drop(columns ='Unnamed: 0', inplace=True)
```

```python
data.head(3)
```

```python
data.isnull().sum()
```

* Now check if the data is lienarly saparable data or non linearly saparable to kno weather to use kernal or not

```python
sns.scatterplot(x='cgpa', y='iq', data=data, hue='placement')
```

* As our dataset is very small so we can easly graph it
* But if it is very large then you need to do hit and trial method and check for both

```python
x =data.iloc[:,:-1]
y = data['placement']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test, y_train, y_test = train_test_split(x,y, train_size=0.8, random_state=10)
```

**Applying SVM**
* As our data is type of classification

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear') #as we have linearly saparable data so usig linear kernal
svc.fit(x_train, y_train)
```

```python
svc.score(x_test, y_test)
```

```python
svc.score(x_train, y_train)
```

**Plotting Decision Region**

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=svc)
```
