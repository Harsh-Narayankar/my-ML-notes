# üîó Correlation & Correlation Coefficients

Correlation is a statistical technique that shows how **strongly** two variables are **related** to each other.

---

## üìò What is Correlation?

- Correlation measures the **direction** and **strength** of a **linear relationship** between two variables.
- It does **not imply causation** ‚Äî just association.

---

## üìà Types of Correlation

| Type                    | Description                                      |
|-------------------------|--------------------------------------------------|
| **Positive Correlation** | As X increases, Y also increases                 |
| **Negative Correlation** | As X increases, Y decreases                      |
| **No Correlation**       | No predictable pattern or linear relationship    |

---

## üßÆ Correlation Coefficient (r)

- It is a **numeric summary** of the relationship.
- Value of `r` lies between **-1 and +1**

### üî¢ Interpretation of r:

| r Value Range       | Strength of Relationship   |
|---------------------|-----------------------------|
| +0.90 to +1.00      | Very strong positive        |
| +0.70 to +0.89      | Strong positive             |
| +0.40 to +0.69      | Moderate positive           |
| +0.10 to +0.39      | Weak positive               |
|  0                  | No correlation              |
| -0.10 to -0.39      | Weak negative               |
| -0.40 to -0.69      | Moderate negative           |
| -0.70 to -0.89      | Strong negative             |
| -0.90 to -1.00      | Very strong negative        |

---

## üîπ Pearson Correlation Coefficient (r)

### ‚úÖ Definition:
- Measures the **strength and direction** of a **linear** relationship.
- Most commonly used.

### üßÆ Formula:
$$
r = \frac{ \sum_{i=1}^n (X_i - \bar{X})(Y_i - \bar{Y}) }{ \sqrt{ \sum_{i=1}^n (X_i - \bar{X})^2 } \cdot \sqrt{ \sum_{i=1}^n (Y_i - \bar{Y})^2 } }
$$

ü§ñ How Correlation is Useful in Machine Learning
1. Feature Selection
In ML, we often have many input variables (features). Correlation helps identify:

Which features are strongly related to the target variable

Which features are highly correlated with each other, which might cause multicollinearity (especially bad for linear models)

üëâ Example:
If two features are 95% correlated, keeping both may add noise. You might drop one to make the model more efficient and generalize better.

2. Model Simplification
Removing correlated or irrelevant variables makes the model:

Faster to train

Easier to interpret

Less prone to overfitting

3. Understanding Feature Impact
Before training a model, correlation gives you a quick diagnostic of:

Which features have predictive potential

Whether a relationship is positive or negative

4. Choosing the Right Model
Some models (like linear regression) assume no multicollinearity. If features are highly correlated:

You may need to use dimensionality reduction (like PCA)

Or choose tree-based models, which are more tolerant of correlation

‚ö†Ô∏è Important Caveat in ML:
Correlation does not imply causation.
In ML, we use correlation as a signal, but we don‚Äôt assume it's the cause. A feature could be strongly correlated with the target without being the real reason.
