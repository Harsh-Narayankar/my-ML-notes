# Hierarchical Clustering
* Used only on linearly saparable
* To group unlabled dataset into clusters
* **Dendogram** : Clsutering is done with tree like structure

* **Dendogram** is a tree like pattern used to store each step as memory that HC algorithm performs 
* x-axis we get number of data points
* y-axis cluster distance (Euclidean)
* 

**2 Approch for Hierarchical clustering**
* This are the ways to create a dendogram  
1) **Agglomerative** : This is a bottom up approch, in which the algorithm stars iwth talikg all data points as single cluster and merge them until one cluster is left
* *Process*
  * It take two nearest point a design a cluster 
  * Then link the remaing nearest points into cluster
  * Until everithing is formed into a single clauster 
 



2) **Divisive** : Divisive algorithm is the reverse of the agglomerative algothim i.e top-down approach
* *Process*
  * First it groups every datapoint a treat it as a single cluster
  * Then breck the cluster in two through maximum distance
  * Till every point is left alone




Most used is Agglomerative cluster

* **Linkage** - Distance between two clusters
* 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"D:\Data Science\Datasets\Iris.csv")
data.drop(columns=['Id', 'Species'], inplace=True)
```

```python
data.head(3)
```

* Cheack if Linear or non Linear as Aglometric clustering only works on linear

```python
sns.pairplot(data)
```

Now we will cheack how many number of cluster can we put through Dendogram

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,10))
sc.dendrogram(sc.linkage(data, method='single', metric='euclidean'))
plt.plot()
```

* The longest line is from 0.8 to 1.6 without cutting
* When we cut a line from middle then we get two number of clusters
* 

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=2, linkage='single')
data['Predicted'] = ac.fit_predict(data)
```

```python
data.head(3)
```

```python
sns.pairplot(data, hue='Predicted')
```
