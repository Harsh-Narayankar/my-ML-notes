# DBSCAN Clustering
* This works on non-linear saparable data
* Density based clustering, ie whereever their is more density , cluster will be created their
* And the left out points are treated as noise points or outliers

* Usind DBSCAN ove K-mean
* 
* Here the k-mean just assumed the center point (red) and divided the dataset into two which is not a proper cluster
* And so DBSCAN clustered those who are more densed

* This is rebust to outliers (ie os not effected by outliers)
* So we can use this to detect outliers, as it makes clusters then the left one are outliers

**How does this makes groups**
1) Minimum points : points inside a circle
2) Epsilon (radius) : radius of circle
3) Core point : center point by which cirlce is design (not included in min points)
4) Boundary point : last left point which are < min point value (and closer to the dense)
5) Noise point : which are away from the dense and not formed a circle

Note : Circle can overlap i.e same points can be shared between two or more clusters


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_moons #this is used to generate a build in dataset
```

```python
x , y = make_moons(n_samples=250, noise=0.1)
```

```python
x.shape
```

```python
df = {'data1': x[:,0],
      'data2': x[:,1]}
```

```python
data = pd.DataFrame(df)
data.head(3)
```

Cheack for non-linear as DBSCAN works on this

```python
sns.scatterplot(x='data1', y='data2', data=data)
plt.show()
```

Now we will use DBSCAN clustering

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.3, min_samples=5)
data['Predicted'] = db.fit_predict(data)
```

```python
data.head(3)
```

```python
sns.scatterplot(x='data1', y='data2', data=data, hue='Predicted')
```

Now check for original

```python
sns.scatterplot(x='data1', y='data2', data=data, hue=y)
```
