# Dropout Layer

**To reduce overfitting**
* We will use dropout layer to reduce overfitting
* We will drop the forward and backward connection of some hidden layer nodes and create a new network architecture
* Why beacause they are the nodes only which will train the model very much and deep and some nodes trains well and some get less information
* As some nodes didnt get trained equally which will result in less accuracy in unknow data

* So we need to drop those nodes ie say 20% of nodes should not work in every epoch
* In each epoch randomly 20% of nodes will not participat
* This results in less dependancy on other nodes ie every nodes get trained well
