# Vanishning Gradient Problem

* While back prapogating to update weightage and bias through gradient decent
* At some place we will see that the weightsvalues are not updating much like 0.01x0.1x0.1 = 0.0001 something like this. This is called Vanishing Gradient problem

* Also the loss is not decresing frequenlty 

**Reason of cause**  
* This is cause because of very deep neural networks ie number of hidden layers are more in the less feature data
* And hidden layer activation are Sigmoid and Tanh cause this give output in the range (0,1)

**How to solve**  
* By not using Sigmoid and Tanh in hidden layers instead we can use ReLu (0, max)
* Weight initialisation should be done by ourself
* Using Batch Normalisation (scaling a data) while the input is passing between hidden layers
