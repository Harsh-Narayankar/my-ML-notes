# SVM in Regression
* For continuous outcome

* At classification we saw that support vectors are taken from inside to get the max distance between those
*  

* Here now in regression we will choise the overmost points as support vectors and minimise the distance
* 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"D:\Data Science\Datasets\placement.csv")
data.head(3)
```

```python
data.isnull().sum()
```

```python
sns.scatterplot(x='cgpa', y='package', data=data)
```

```python
x = data[['cgpa']] #We need to pass x as 2D
y= data['package']
```

* We will apply SVM on this

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train , y_test = train_test_split(x,y, train_size=0.8, random_state=42)
```

```python
from sklearn.svm import SVR #support vector regressor
```

```python
svr = SVR()
svr.fit(x_train, y_train)
```

```python
print('For train: ',svr.score(x_train, y_train))
print('Fot test ', svr.score(x_test, y_test))
```

But as our dataset is like Linear so 

```python
svr_linear = SVR(kernel='linear')
svr_linear.fit(x_train, y_train)
```

```python
print('For train: ',svr_linear.score(x_train, y_train))
print('Fot test ', svr_linear.score(x_test, y_test))
```

**Plotting**

```python
sns.scatterplot(x='cgpa', y='package', data=data)
plt.plot(data['cgpa'],svr_linear.predict(x), color='r')
```
