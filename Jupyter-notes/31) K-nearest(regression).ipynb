```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
data = pd.read_csv(r"D:\Data Science\Datasets\Salary_Data.csv")
data.head(10)
```

* So here we will pass YearsExperience and Age then we need to predict the Salary
* It falls under Regression because the salary is contineous in nature

```python
data.isnull().sum()
```

```python
x = data.iloc[:,:-1]
y = data['Salary']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc.fit(x)
```

```python
x = pd.DataFrame(sc.transform(x), columns = x.columns)
```

```python
x.head(2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.8, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn_reg = KNeighborsRegressor()
knn_reg.fit(x_train, y_train)
```

```python
knn_reg.score(x_test, y_test)
```

```python
knn_reg.score(x_train, y_train)
```
