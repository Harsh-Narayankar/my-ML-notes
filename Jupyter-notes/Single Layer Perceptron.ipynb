# Single Layer Perceptron
* This is also called perceptron
* Base neural network used for CNN and ANN 

* We get a input *xi's* which is then multiplyed with *w* (Every connection between two neurons has a weight, which tells the network how strongly one neuron influences another)
* Then add a *bias* term

* We know the simple linear equation is y=mx+c (were: m is slope of curve and c in intercept)
* When we use same equation for multiple data then the equation becomes y=w1x1+w2x2+....+wnxn + b
* Now all the data will be passed to activation function (for desired output ie classification or regression)
* Then we will get the output
* 
*

* 
* The actual equation looks like this
* Were:
  * wT is a row vector (1x3)
  * x is a columns vector (3x1)
  * Both dot product (gives scaler single value)
  * And the resultant is 1x1 which is added to the bias

**Not getting desired output**
* If we are not getting desired output then we do back prapogation and change w (weightage) and b (bias) values

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
data = pd.read_csv(r"D:\Data Science\Datasets\placement-dataset.csv")
data.drop(columns='Unnamed: 0', inplace=True)
```

```python
data.head(3)
```

We will check how does the data looks

```python
sns.scatterplot(x='cgpa', y='iq', hue='placement', data=data)
```

```python
x = data.iloc[:,:-1]
y = data['placement']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.8, random_state=10)
```

Calling perceptron

```python
from sklearn.linear_model import Perceptron
```

```python
pr = Perceptron()
pr.fit(x_train, y_train)
```

```python
pr.score(x_test, y_test), pr.score(x_train, y_train)
```

graphical visuals

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=pr)
```
