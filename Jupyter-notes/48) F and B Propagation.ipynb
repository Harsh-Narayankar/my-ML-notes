# Back prapogation
* This is the method of updating weightage and bias value to imporve the accuracy. Beacause this is the main reason we are not geeting our desired output

* In many cases the wights we initialised at the bigining are not accurate so e need to see its output and back propogate and update the values to decrease the error

* The backproportion uses Gradient decendent

**Gradient Decent**
* Go back and update the weights values
* So **W_new = W_old - lamda(dL/dW)**
  * were :
    * W_new - updated value of W after back prapogation
    * W_old - old value of W which was not giving desired output
    * lamda - learning rate (HP) 
    * L - loose function
    * W - wight
  * If we are using regression data then we will use **MSE = 1/N(sum(y_og-y'))^2**
  * Root MSE - in original unit
  * Mean Absolute Error - less sensitive to outlier
