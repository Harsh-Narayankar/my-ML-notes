```python
df = pd.DataFrame({'name': ['cat','dog','black','honey']})
df
```

So we have a data, and now we need to do encoding of it, for that we may require sklearn library

```python
from sklearn.preprocessing import LabelEncoder
```

fit_transform() is a common method used in data preprocessing, especially in machine learning workflows using libraries like scikit-learn.

üîç Purpose of fit_transform():
It combines two steps in one:

fit() ‚Äì Learns the parameters/statistics from the data (e.g., mean and std for scaling).

transform() ‚Äì Applies the learned transformation to the data.

‚úÖ So fit_transform() is used to:
Learn from the data and immediately apply the transformation.

üîß Why use fit() and transform() separately?  
1. Training vs. Test Data (Prevent Data Leakage)  
You must fit on training data only, then transform both training and test data using the same learned parameters.

```python
le = LabelEncoder()
# fit: will train the model
# transform: will convert the model (combining both will used for training model)
le.fit_transform(df[['name']])
```

```python
# so now we need to attach this encodel column to our column
le.fit(df)
```

```python
df['encoded value'] = le.fit_transform(df)
```
