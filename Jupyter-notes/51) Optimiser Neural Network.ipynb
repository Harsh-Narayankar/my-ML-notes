# Optimiser Neural Network

* When we didnt get the desired output and getting lot of Loss
* Then to reduce the loss we do back prapogation and use gradient desend and update the weight
* Gradient desend works on Loss function which has to be in parabolic shape and goes towards the global minima

* But if the function between loss and weight, bias is something like thi then ?
* Where we have lot of minimum points were we have local and global minima
* GD problem is it update wieght step by step by single learing rate
* But if it reaches the local minima then it will struck their and we will not reach the global minima
* So to help in this situaltion here comes a **Optimizer**

**Optimiser**
* Used when the curvature has lost of local mininma (not the lowest)
* Here the learning rate values gets updated throught the process according to our data
