```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"D:\Data Science\Datasets\Salary_Data.csv")
data.head(3)
```

We ill try decision tree on this on Ae , experisnace and salary

```python
data.isnull().sum()
```

```python
sns.pairplot(data)
```

* This is a linear in nature but yet we will apply DT on this

```python
x = data.iloc[:,:-1]
y = data['Salary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test , y_train , y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

* Here in the criterion you will have same as regression i.e squared_error, friedman_mse

```python
dtr = DecisionTreeRegressor(criterion='squared_error')
dtr.fit(x_train, y_train)
```

```python
dtr.score(x_train, y_train)
```

```python
dtr.score(x_test, y_test)
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(dtr)
plt.show()
```

**Improve Accuracy**

```python
# Pre-pruning
dtr1 = DecisionTreeRegressor(max_depth=5)
dtr1.fit(x_train, y_train)
```

```python
dtr1.score(x_train, y_train)
```

```python
dtr1.score(x_test, y_test)
```

**Post-pruning**

```python
for i in range(1,20):
    dtr2 = DecisionTreeRegressor(max_depth=i)
    dtr2.fit(x_train, y_train)
    print('Train score: ',dtr2.score(x_train, y_train), 'Test score: ',dtr2.score(x_test, y_test), ' in depth: ',i)
```
