# Stratified K-Fold Cross-Validation
* This works on Unbalanced data
* **Unbalanced** we could find this in classfication data, i.e say we have 3 classes but the number of values in each calss in not same, i.e somea calss have 100 values while other have 50 or 200
* 

* So when we apply Stratified k-fold then it makes sure that all the class values are in validation part
* 

* So **Stratified k-Fold Cross-Validation** means the dataset is equally partetioned into k subparts
* Iteratively
*  Out of k subpart 1 part is selected as validation and k-1 as training data
*  Where **Stratified** solves the problem of imbalanced data

# Leave-one-out Cross-Validation

**See How CV methods works**

```python
new_data = data.head(10)
```

```python
x_new = new_data.iloc[:,:-1]
y_new = new_data['package']
```

```python
from sklearn.model_selection import LeaveOneOut, LeavePOut, KFold, StratifiedGroupKFold
```

**Leave one out** leave one data iteratively for testing and rest for training

```python
loouCV = LeaveOneOut()

for train, test in loouCV.split(x_new,y_new):
    print(train, test)
```

```python
lpuCV = LeavePOut(p=2)

for train, test in lpuCV.split(x_new,y_new):
    print(train, test)
```

```python
k_fold = KFold(n_splits=5)

for train, test in k_fold.split(x_new, y_new):
    print(train, test)
```

```python
stra_kfold = StratifiedGroupKFold(n_splits=5)

for train, test in stra_kfold.split(x_new,y_new):
    print(train, test)
```
