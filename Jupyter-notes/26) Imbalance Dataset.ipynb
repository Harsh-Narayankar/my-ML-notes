# Imbalanced Dataset
* Say we have a dataset with may category and if once category is frequenly more then it is an Imbalanced dataset
* So if  we train the model with high values of some category then the model will be baised towards that

* that is if we train 500 rows of cat in 600 rows dataset then the model will be bised towards cat

**Two Techniques**
1) Random Under Sampling : It will takes some subset of large category data and make it equal to the lowest catogory data (large to small)
2) Random Over Sampling : We will increses the minority classes by repetation (duplicasy) till it matches to mejority (small to large)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"D:\Data Science\Datasets\Social_Network_Ads.csv")
data.drop(columns = ['User ID', 'Gender'] , inplace=True)
```

```python
data.head(3)
```

```python
data['Purchased'].value_counts()
```

So as value count of 0 is more therefore our model will be bised towards 0

```python
x = data.iloc[:,:-1]
y = data['Purchased']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train, y_test = train_test_split(x,y, train_size=0.8, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

Check for biaseness through iteration over the values where y_actual is 1 but you got 0

* So now we need to balance the data through a module called **imblearn**

```python
from imblearn.under_sampling import RandomUnderSampler 
```

* *RandomUnderSampler* returns 4 data i.e take two data input and output and gives 4 balanced data

```python
x = data.iloc[:,:-1]
y = data['Purchased']
```

```python
ru = RandomUnderSampler()
ru_x, ru_y = ru.fit_resample(x,y)
```

```python
ru_x
```

```python
ru_y
```

* Check if our data has beed resampled through undersampling

```python
ru_y.value_counts()
```

So our new input(x) and output(y) are ru_x and ru_y

Now training model again with resampled data

```python
x_train, x_test,y_train, y_test = train_test_split(ru_x,ru_y, train_size=0.8, random_state=42)
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

# Random OverSampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
x = data.iloc[:,:-1]
y = data['Purchased']
```

```python
ro = RandomOverSampler()
ro_x, ro_y = ro.fit_resample(x,y)
```

```python
ro_y.value_counts()
```

```python
x_train, x_test,y_train, y_test = train_test_split(ro_x,ro_y, train_size=0.8, random_state=42)
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```
