# Linear Regression vs Logistic Regression

##  1. Purpose

| Aspect | Linear Regression | Logistic Regression |
|--------|-------------------|---------------------|
| Goal   | Predict **continuous values** (e.g., price, salary) | Predict **categorical outcomes** (e.g., yes/no, spam/not spam) |

---

##  2. Type of Output

| Linear Regression | Logistic Regression |
|------------------|---------------------|
| Outputs a real number (e.g., 2.7, 100.5) | Outputs a **probability** (between 0 and 1), then classifies using a threshold (default: 0.5) |

---

##  3. Mathematical Function

| Linear | Logistic |
|--------|----------|
| `y =  + x + x + ... + x` | `P(y=1) = 1 / (1 + e^-( + x + ... + x))` (Sigmoid function) |

---

##  4. Use Case Examples

| Linear Regression | Logistic Regression |
|------------------|---------------------|
| Predicting house prices, temperature, sales | Email spam detection, loan approval (yes/no), disease prediction (positive/negative) |

---

##  5. Error Measurement

| Linear | Logistic |
|--------|----------|
| Uses **Mean Squared Error (MSE)** | Uses **Log Loss (Cross-Entropy Loss)** |

---

##  6. Output Interpretation

| Linear Regression | Logistic Regression |
|------------------|---------------------|
| Coefficients show how much the dependent variable changes per unit change in predictor | Coefficients show effect on the **log-odds** of the outcome |

---

##  7. Assumptions

| Linear Regression | Logistic Regression |
|------------------|---------------------|
| Linear relationship between variables | Logistic (sigmoid) relationship |
| Normal distribution of errors | Binary/categorical output; errors dont need to be normal |
| Homoscedasticity (constant variance) | No such requirement |

---

##  8. Algorithm Type

| Linear | Logistic |
|--------|----------|
| **Regression** algorithm | **Classification** algorithm |

---

##  When to Use Which?

| Situation | Use |
|-----------|-----|
| Predicting a **quantity** | Linear Regression |
| Predicting a **class/category** | Logistic Regression |

**OVR** Method

The difference between ovr (One-vs-Rest) and multinomial lies in how multi-class classification is handled, especially in models like logistic regression.

Lets break it down:

 1. ovr  One-vs-Rest (One-vs-All)
The model builds one binary classifier per class.

 How it works:
For N classes, it trains N binary classifiers.

Each classifier separates one class vs the rest.

During prediction, the class with the highest probability is chosen.

 Used in:
LogisticRegression(multi_class='ovr') (in scikit-learn)

 Pros:
Simple and easy to implement

Works even if model doesnt support multiclass natively

 Cons:
May produce inconsistent probabilities

Not ideal for correlated or overlapping classes

 2. multinomial  Multinomial Logistic Regression (Softmax)
The model builds a single model that predicts all classes at once using a softmax function.

 How it works:
Learns jointly for all classes.

Outputs a probability distribution over all classes.

Uses cross-entropy loss for optimization.

 Used in:
LogisticRegression(multi_class='multinomial', solver='lbfgs')

 Pros:
Mathematically correct for multi-class problems

Probabilities are consistent and sum to 1

Often gives better performance

 Cons:
Needs solvers that support it (lbfgs, sag, saga, newton-cg)

Slightly more complex to train

 Summary Table:
Feature	ovr (One-vs-Rest)	multinomial (Softmax)
Models Trained	One per class	One for all classes
Output	Best of binary predictions	Softmax probabilities
Uses binary base?	Yes	No (direct multiclass)
Performance	Usually lower	Usually higher
Preferred for	Simple or binary-only models	True multiclass problems
