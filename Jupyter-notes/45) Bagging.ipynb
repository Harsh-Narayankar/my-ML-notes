# Bagging (or bootstrap agrigation)
* This is a parellel technique to improve our model
* means, it will breck you original data into small small part i.e train data will be brecked
* Note the subset should be less the train set
* 
* **Bagging** here means this subsets is given to model and all the models are same
* i.e To the same model we will give multiple subset
* 

Note: All the model are same ex DT but it should be DT1, DT2, DT3....
* So we will combine the output and make a single output

* Whenever test data comes, we will load the test data(same) into all the models dt1,dt2...
* And combine the output into single

* If classification : MODE
* If Regression : AVG

**Summery**    
Models should be same but called differently  
While training we will load sets of data respectively and while new data comes we will load same data into all the model

# Random Forest
* This is part of DT ie base model is DT and we cannot change it
* This also follows the Bagging technique

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=1000, noise=0.2)
```

```python
df = {'x1': x[:,0], 'x2': x[:,1], 'y':y }
```

```python
data = pd.DataFrame(df)
```

```python
sns.scatterplot(x='x1', y='x2', hue='y', data=data)
```

```python
x_a = data.iloc[:,:-1]
y_a = data['y']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test, y_train , y_test = train_test_split(x_a,y_a, train_size=0.8, random_state=42)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
bg = BaggingClassifier(estimator=SVC(), n_estimators=30)
bg.fit(x_train, y_train)
```

```python
bg.score(x_train, y_train), bg.score(x_test, y_test) 
```

```python
rf = RandomForestClassifier(n_estimators = 30)
rf.fit(x_train, y_train)
```

```python
rf.score(x_test,y_test) , rf.score(x_train , y_train)
```
